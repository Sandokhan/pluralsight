```
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from matplotlib.path import Path

plt.style.use('seaborn-white')
```

```
fig, ax = plt.subplots()

ax.add_patch(
   patches.Rectangle(
   (.1, .1),
    .5,
    .5,
    fill=False)
)

plt.show()
```

```
fig, ax = plt.subplots()

ax.add_patch(
   patches.Rectangle(
   (.1, .1),
    .5,
    .5,
    facecolor='yellow',
    edgecolor='green')
)

plt.show()
```

```
fig, ax = plt.subplots()
ax.set_aspect(aspect='equal')

for p in [
    patches.Rectangle((.1,.1), .3 ,.6,
                     hatch='.'),
    patches.Rectangle((.5,.1), .3 ,.6,
                     hatch='\\',
                     fill=False),
]:
    ax.add_patch(p)
    
plt.show()
```

```
fig, ax = plt.subplots()

p = patches.PathPatch(Path([ (0.1, 0.1), (0.8, 0.8), (0.8, 0.1), (0.4, 0.2)],
                          [Path.MOVETO, Path.LINETO, Path.LINETO, Path.CLOSEPOLY]),
                     fill=None)
ax.add_patch(p)

plt.show()
```

```
fig, ax = plt.subplots()

p = patches.PathPatch(Path([ (0.1, 0.1), (0.8, 0.8), (0.8, 0.1), (0.4, 0.2)],
                          [Path.MOVETO, Path.LINETO, Path.CURVE3, Path.CURVE3]),
                     fill=None)
ax.add_patch(p)

plt.show()
```
