```
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from matplotlib.path import Path
import numpy as np
import pandas as pd

plt.style.use('seaborn-white')
```

```
fig, ax = plt.subplots()

ax.add_patch(
   patches.Rectangle(
   (.1, .1),
    .5,
    .5,
    fill=False)
)

plt.show()
```

```
fig, ax = plt.subplots()

ax.add_patch(
   patches.Rectangle(
   (.1, .1),
    .5,
    .5,
    facecolor='yellow',
    edgecolor='green')
)

plt.show()
```

```
fig, ax = plt.subplots()
ax.set_aspect(aspect='equal')

for p in [
    patches.Rectangle((.1,.1), .3 ,.6,
                     hatch='.'),
    patches.Rectangle((.5,.1), .3 ,.6,
                     hatch='\\',
                     fill=False),
]:
    ax.add_patch(p)
    
plt.show()
```

```
fig, ax = plt.subplots()

p = patches.PathPatch(Path([ (0.1, 0.1), (0.8, 0.8), (0.8, 0.1), (0.4, 0.2)],
                          [Path.MOVETO, Path.LINETO, Path.LINETO, Path.CLOSEPOLY]),
                     fill=None)
ax.add_patch(p)

plt.show()
```

```
fig, ax = plt.subplots()

p = patches.PathPatch(Path([ (0.1, 0.1), (0.8, 0.8), (0.8, 0.1), (0.4, 0.2)],
                          [Path.MOVETO, Path.LINETO, Path.CURVE3, Path.CURVE3]),
                     fill=None)
ax.add_patch(p)

plt.show()
```

```
fig, ax = plt.subplots()
ax.plot([1,2,3],
        [2,4,6])
plt.show()
```

```
fig, ax = plt.subplots()
ax.plot([1,2,3],
        [2,4,6])

ax.annotate('min value',
            xy=(1,2), #point which will be highlighted
            xytext=(1.5,2.0), #coordinates of bottom left of the text annotation
            arrowprops=dict(color='g')
           )

plt.show()
```

```
fig, ax = plt.subplots()
ax.plot([1,2,3],
        [2,4,6])

ax.annotate('min value',
            xy=(1,2), #point which will be highlighted
            xytext=(1,3), #coordinates of bottom left of the text annotation
            arrowprops=dict(facecolor='y', edgecolor='green', alpha=.3)
           )

plt.show()
```

```
fig, ax = plt.subplots()
ax.plot([1,2,3],
        [2,4,6])

ax.annotate('min value',
            xy=(2, 4), #point which will be highlighted
            xytext=(2.0,2.5), #coordinates of bottom left of the text annotation
            arrowprops=dict(color='g', shrink=.1)
           )

ax.plot([2], [4], 'go') #Highlight the point of interest
plt.show()
```

```
np.random.seed(0)

x1 = -1 + np.random.randn(100)
y1 = -1 + np.random.randn(100)

x2 = 1 + np.random.randn(100)
y2 = 1 + np.random.randn(100)
```

```
fig, ax = plt.subplots()

ax.scatter(x1,y1, color='r')
ax.scatter(x2,y2, color='g')

bbox_props = dict(boxstyle='square',
                  facecolor='w',
                  alpha=.5)

ax.text(-2, -2,
        'Sample A', 
        ha='center',
        va='center',
        size=20, 
        bbox=bbox_props)

ax.text(2, 2,
        'Sample B', 
        ha='center',
        va='center',
        size=20, 
        bbox=bbox_props)
```
